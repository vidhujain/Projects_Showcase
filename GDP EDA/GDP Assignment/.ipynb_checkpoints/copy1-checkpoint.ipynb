## Feature Engineering

### Lets create new column by combining Items  description and duration column

```python
niti_df['Items  Description'].value_counts()
```

```python
# niti_df1['Items  Description']=niti_df1[niti_df1['Items  Description'].
#                                         str.contains('GSDP')]['Items  Description'
#                                                              ].str.replace("GSDP - CURRENT PRICES (` in Crore)",
#                                                                          'GSDP_CURRENT_PRICES_CR',
#                                                                         regex=False)
niti_df['Items  Description'].replace("GSDP - CURRENT PRICES (` in Crore)",
                                                                         'GSDP_CURRENT_PRICES_CR',
                                                                        regex=False,inplace=True)
niti_df['Items  Description'].replace("(% Growth over previous year)",'Growth_Perc_Over_PrevYr',
                                                                        regex=False,inplace=True)
```

```python
niti_df['Items_Duration'] = niti_df['Items  Description']+'_'+niti_df['Duration']
niti_df.drop(columns=['Items  Description','Duration'],inplace=True)
## rearrange the columns
col_list = list(niti_df.columns)
#print(col_list)
col_list = [col_list[-1]] + col_list[:-1]
#print(col_list)
niti_df = niti_df[col_list]
niti_df
```

### Lets create another column that distinguishes whether the numbers are for GSDP or for Growth Perc

```python
niti_df['Items'] = niti_df['Items_Duration'].apply(lambda x: 1 if 'GSDP' in x else 0)
## rearrange the columns
col_list = list(niti_df.columns)
#print(col_list)
col_list = [col_list[-1]] + col_list[:-1]
#print(col_list)
niti_df = niti_df[col_list]
niti_df
niti_df
```

### Lets create two different dataframes 
1. DataFrame for GSDP for different years for all states
2. DataFrame for Growth Percentage over different years for all states

```python
gsdp_df = niti_df[niti_df['Items']==1]
growth_df = niti_df[niti_df['Items']==0]
```
